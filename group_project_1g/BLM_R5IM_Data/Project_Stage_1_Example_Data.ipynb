Instead of running the MQTT broker and overloading the network, we will use stored data given here. You may have to investigate methods of providing / changing this data over time. Use the python time library for this (the sleep function may be useful)

This work may be useful for group 1E, however there may be better ways of reading / writing to file.

```python
import glob
import pandas as pd
#import random
#import paho.mqtt.client as mqtt
import string
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = glob.glob("BLM_R5IM_Data/cycle" + '/*.csv')
```

```python
for file in files:
    print(file)
```

```python
selected_file = files[0]
```

```python
input_data = pd.read_csv(selected_file)
dataframe = input_data.drop(columns = input_data.columns[0]).to_numpy()
```

```python
dataframe
```

```python
x_data = np.linspace(-0.5, 10.5, 2200)
```

```python
plt.scatter(x_data, dataframe[39]);
plt.title('R5IM');
plt.xlabel('Time [ms]');
plt.ylabel('Signal [V]');
plt.grid()
```

```python
plt.scatter(x_data, dataframe[38]);
plt.title('R9BLM4');
plt.xlabel('Time [ms]');
plt.ylabel('Signal [V]');
plt.grid()
```
