```python
import math
from scipy.integrate import quad

def integrand(x):
    return 1/x**2

for i in range(1,13):
    n = 10**i
    I = quad(integrand, 1.45137, n)
    print((',n,'),' = ' ,I[0], '  Error bound = ', I[1], sep = "")
```

```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, num=1000)
y = [1 for i in range(len(x))]
plt.plot(x,y)
y_int = integrate.cumulative_trapezoid(y, x)
plt.plot(x[:-1], y_int)

plt.show()
```
